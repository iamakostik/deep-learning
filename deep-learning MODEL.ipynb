```python
#import standard libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load the dataset
df = pd.read_csv('fundamentals.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0','Ticker Symbol','Period Ending','Cash Ratio','Current Ratio','Quick Ratio', 'Estimated Shares Outstanding', 'For Year','Net Income'])
```

```python
df.head()
```

```python
df.isna().sum().to_string()
```

```python
len(df.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
df.shape
```

```python
fig, axs = plt.subplots(17,5, figsize=(30,80),)
fig.subplots_adjust(hspace = 0.5, wspace=.2)
axs = axs.ravel()

for index, column in enumerate(df.columns):
    axs[index-1].set_title("{} vs. Earnings Per Share".format(column),fontsize=19)
    axs[index-1].scatter(x=df[column],y=df['Earnings Per Share'],color='blue',edgecolor='k')
    
fig.tight_layout(pad=1)
```

```python
features_df = df.drop(columns=['Earnings Per Share'])
target = df['Earnings Per Share']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, target,
test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
```

```python
#Get the predictions
predicted_values = linear_model.predict(x_test)
```

```python
round(linear_model.score(x_test, y_test),3)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
#RSS
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#R Squared
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
#Ridge regression

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
#compare weights

def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
```

```python
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
